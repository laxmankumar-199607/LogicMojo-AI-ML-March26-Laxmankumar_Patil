```python
import re

data = [
    "User: Mahipal | Phone: 9876543210 | Email: mahi@gmail.com | Amount: Rs.5000",
    "User: Ravi | Phone: 98765abc10 | Email: ravi#gmail.com | Amount: USD300",
    "User: NULL | Phone: 9123456789 | Email: test@yahoo.com | Amount: Rs.7000",
    "User: Ankit | Phone: 9999999999 | Email: ankit@gmail.com | Amount: Rs.0",
    "User: Sita | Phone: 8888888888 | Email: sita@gmail.com | Amount: Rs.-100"
]

clean_data = []
errors = []
# 1.HANDLE EXCEPTIONS
for record in data:
    try:
        # Extract using regex
        name_match = re.findall(r"User:\s*(\w+)", record)
        phone_match = re.findall(r"Phone:\s*(\d{10})", record)
        email_match = re.findall(r"Email:\s*([\w\.-]+@[\w\.-]+)", record)
        amount_match = re.findall(r"Amount:\s*(?:Rs|USD)\.?(-?\d+)", record)
        currency_match = re.findall(r"Amount:\s*(Rs|USD)", record)

        if not (name_match and phone_match and email_match and amount_match and currency_match):
            raise ValueError("Invalid format")

        name = name_match[0]
        phone = phone_match[0]
        email = email_match[0]
        amount = int(amount_match[0])
        currency = currency_match[0]

        #  2. USING ASSERT
        assert name.lower() != "null", "Invalid name"
        assert amount > 0, "Amount must be greater than 0"

        #  3. REGEX EXTRACTION
        name = re.sub(r'[^a-zA-Z]', '', name).lower()
        email = email.lower()
        currency = currency.lower()
        phone = re.findall("\d{10}", phone)

        clean_data.append({
            "name": name,
            "phone": phone,
            "email": email,
            "amount": amount,
            "currency": currency
        })

    except Exception as e:
        errors.append(f"Error in record: {record} --> {str(e)}")

print("Clean Data:")
for item in clean_data:
    print(item)

print("\nErrors:")
for err in errors:
    print(err)
```
