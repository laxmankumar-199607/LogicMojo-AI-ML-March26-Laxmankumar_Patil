```python

## Question 1:- A digital payment platform recorded the following transaction amounts made by a customer during a single day.
# 1.1 Identify and display all transaction values that appear more than once.
transactions = [1200, 3000, 250, 7800, 540, 3000, 9100, 250, 1500, 7800]
more_than_once_trnx = []
for trxn in transactions:
    if transactions.count(trxn) > 1:
        more_than_once_trnx.append(trxn)
        transactions.remove(trxn)

print(more_than_once_trnx)
#print(transactions)
```

```python
## 1.2 Detect transactions greater than ₹5000 as these may require fraud monitoring.
transactions = [1200, 3000, 250, 7800, 540, 3000, 9100, 250, 1500, 7800]
fraud_monitor = []
for t in transactions:
    if t > 5000:
        fraud_monitor.append(t)

print("These transaction requires fraud monitoring: " , fraud_monitor)
```

```python
## 1.3 Calculate the total spending for the day.
# transactions = [1200, 3000, 250, 7800, 540, 3000, 9100, 250, 1500, 7800]
total_spending = sum(transactions)
print("Total spending for the day is:",total_spending)
```

```python
## 1.4 Create a new list containing only unique transaction values.
transactions = [1200, 3000, 250, 7800, 540, 3000, 9100, 250, 1500, 7800]
for transaction in transactions:
    if transactions.count(transaction) > 1:
        transactions.remove(transaction)

print(transactions)
```

```python
## 1.5 Sort the unique transaction list in descending order to view the largest transactions first.
transactions = [1200, 3000, 250, 7800, 540, 3000, 9100, 250, 1500, 7800]
for trxn in transactions:
    if transactions.count(trxn) > 1:
        transactions.remove(trxn)

transactions.sort(reverse=True)
print("Unique transactions in descending order: ", transactions)
```

```python
##Question 2:- A dataset exported from a database contains the following customer age values.

# ages = [25, 42, -1, 34, None, 52, 17, -5, 29, None, 46] Some entries are invalid because they are either negative numbers or missing values (None)

## 2.1 Iterates through the dataset and removes all invalid age values.

ages = [25, 42, -1, 34, None, 52, 17, -5, 29, None, 46]
for age in ages:
    if age is None or age < 0:
       ages.remove(age)

print("After removed all invalid age values: ",ages)
```

```python
## 2.2 Stores only valid ages in a new list.

ages = [25, 42, -1, 34, None, 52, 17, -5, 29, None, 46]
valid_age_data = []
for age in ages:
    if age is not None and age > 0:
        valid_age_data.append(age)

print("Valid ages in a new list: ",valid_age_data)
```

```python
## 2.3 Calculates the average age of valid customers.

ages = [25, 42, -1, 34, None, 52, 17, -5, 29, None, 46]
valid_age_data = []
for age in ages:
    if age is not None and age > 0:
        valid_age_data.append(age)

# print(valid_age_data, len(valid_age_data))
print("The average age of valid customers:", sum(valid_age_data) / len(valid_age_data))
```

```python
## 2.4 Displays the list of customers whose age is greater than 30.
ages = [25, 42, -1, 34, None, 52, 17, -5, 29, None, 46]

customer_age_greater_thirty = []
for age in ages:
    if age is not None and age > 30:
        customer_age_greater_thirty.append(age)

print("List of customer whose age is greater than 30: ",customer_age_greater_thirty)
```

```python
## Que 4. A web platform records the hours when users log into the system.

## 4.1 Separates login hours into even-hour logins and odd-hour logins.
login_hours = [1, 3, 5, 7, 2, 6, 8, 4, 9, 10]
even_hours = []
odd_hours = []
for hour in login_hours:
    if hour % 2 == 0:
        even_hours.append(hour)
    else:
        odd_hours.append(hour)

print("Even login hours: ",even_hours)
print("Odd login hours: ",odd_hours)
```

```python
## 4.2 Identifies login hours that fall into peak activity time (hours greater than 5).
login_hours = [1, 3, 5, 7, 2, 6, 8, 4, 9, 10]
peak_activity_hours = []
for hour in login_hours:
    if hour > 5:
        peak_activity_hours.append(hour)

print(peak_activity_hours)
```

```python
## 4.3 Sorts the login hours in ascending order.
login_hours = [1, 3, 5, 7, 2, 6, 8, 4, 9, 10]
login_hours.sort()
print("Sorted the login hours in ascending order:",login_hours)
```

```python
## 4.4 Calculates how many logins occurred during peak hours.
login_hours = [1, 3, 5, 7, 2, 6, 8, 4, 9, 10]
count = 0
for hour in login_hours:
    if 6 <= hour <= 10:
        count += 1

print("Peak hour count:",count)
```

```python
## 5.7 Identify the Most Sold Product Based on the frequency of orders, determine which product appears the most in the dataset.
orders = [ ["ORD001","Laptop",75000,"Completed"],
           ["ORD002","Mouse",500,"Completed"],
           ["ORD003","Laptop",75000,"Returned"],
           ["ORD004","Keyboard",1500,"Completed"],
           ["ORD005","Laptop",75000,"Completed"],
           ["ORD006","Monitor",20000,"Fraud"],
           ["ORD007","Mouse",500,"Completed"],
           ["ORD008","Laptop",75000,"Completed"] ]
freq_list = []
processed_data = []
for i in range(len(orders)):
    product_name = orders[i][1]
    frequency_order = 0
    if product_name in processed_data:
        continue
    for j in range(len(orders)):
        if orders[j][1] == product_name:
            frequency_order+=1

    row = {'Product_Name':product_name,
           'Frequency_Order':frequency_order}
    processed_data.append(product_name)
    freq_list.append(row)

max_product = None
max_count = 0

for item in freq_list:
    if item['Frequency_Order'] > max_count:
        max_count = item['Frequency_Order']
        max_product = item['Product_Name']

print("Most Sold Product:", max_product)
print("Number of Orders:", max_count)
```
